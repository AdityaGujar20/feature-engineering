# Feature Scaling:
* It is a technique to standardize the independent features present in the data in the fixed range.
* Features Scaling is when we want to perform distance based ML algorithms.

<hr>

## Types of Feature Scaling:
1. Standardization (Z-Score Normalization)
2. Normalization (MinMax Scaling)

<hr>

## Standardization (Z-Score Normalization):
* The data will be scaled between -1 to 1.
* The mean and the standard deviation of the scaled column will become 0 and 1 respectively.
* Z-score normalization centers the data around the mean with a unit standard deviation. While outliers do influence the mean and standard deviation, their impact is less dramatic compared to min-max scaling. This is because the z-score measures how many standard deviations a data point is from the mean, which mitigates the effect of extreme values.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/Social_Network_Ads.csv")
df = df.iloc[:,2:]
df.sample(5)
```

```python
# Splitting the data into independent and dependent features x and y respectively
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# split the data into train and test before performing standardization
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# checking the shape of train and test data
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

# creating a scaler object
scaler = StandardScaler()

# fit method is used to learn the patterns from the data
scaler.fit(X_train)

# transform method is used to apply the transformation on the data based on the learnings from the pattern
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# NOTE: the transform method will return the numpy array
```

```python
# mean of transformed Age and EstimatedSalary features 
scaler.mean_
```

```python
# converting the numpy array into the dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns= X_test.columns)
```

## Effects of Standardization

```python
print("Before Scaling")
np.round(X_train.describe(), 1)
```

```python
print("After Scaling")
np.round(X_train_scaled.describe(), 1)
```

```python
# scatter plot before and after applying standardization

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title("Before Scaling")
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])

ax2.set_title("After Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color = 'red')

plt.show()
```

```python
# KDE plot also called Kernel Density Estimate
# KDE plot takes each individual value and draws a gaussian bell curve over it

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Age'], ax = ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax = ax1)

ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Age'], ax = ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax = ax2)

plt.show()
```

## Normalization (MinMAX Scaling)
* The data will be scaled between 0 to 1.
* The min and max value of the scaled column will be 0 and 1 respectively.
* Normalization is highly sensitive towards outliers because min-max scaling relies on the minimum and maximum values, the presence of outliers can significantly affect the scaling process. An outlier with a very large or very small value will stretch the range, compressing the values of the majority of the data points into a narrower interval, making the transformation less effective for these points.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/wine_data.csv", header = None, usecols=[0,1,2])
df.columns = ["Class label", "Alcohol", "Malic Acid"]
```

```python
df.sample(5)
```

```python
color_dict = {
    1: 'red',
    2: 'green',
    3: 'blue'
}

sns.scatterplot(x='Alcohol', y='Malic Acid', hue='Class label', palette=color_dict, data=df)
```

```python
# Splitting the data into independent and dependent features x and y respectively
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
# split the data into train and test before performing normalization
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# checking the shape of the train and test data
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

# creating a scaler object
scaler = MinMaxScaler()

# fit method is used to learn the patterns from the data
scaler.fit(X_train)

# transform method is used to apply the transformation on the data based on the learnings from the pattern
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# NOTE: the transform method will return the numpy array
```

```python
# converting the numpy array into the dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

## Effects of Scaling

```python
print("Before Scaling")
np.round(X_train.describe(), 1)
```

```python
print("After Scaling")
np.round(X_train_scaled.describe(), 1)
```

```python
# scatter plot before and after applying Normalization

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title("Before Scaling")
ax1.scatter(X_train['Alcohol'], X_train['Malic Acid'])

ax2.set_title("After Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic Acid'], color = 'red')

plt.show()
```

```python
# KDE plot also called Kernel Density Estimate
# KDE plot takes each individual value and draws a gaussian bell curve over it

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12, 5))

ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax = ax1)
sns.kdeplot(X_train['Malic Acid'], ax = ax1)

ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'], ax = ax2)
sns.kdeplot(X_train_scaled['Malic Acid'], ax = ax2)

plt.show()
```
